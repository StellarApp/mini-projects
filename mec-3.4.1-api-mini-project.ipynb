### 3. Calculate what the highest and lowest opening prices were for the stock in this period.


```python
highest_open = max(df_dict['Open'].keys(), key=(lambda k: df_dict['Open'][k]))
lowest_open = min(df_dict['Open'].keys(), key=(lambda k: df_dict['Open'][k]))
```

```python
print('the highest open prices in 2017 is ', df_dict['Open'][highest_open], 
      ' and the lowest open prices in 2017 is ', df_dict['Open'][lowest_open])
```

## Search the coincident date for the highest & lowest opening prices

```python
for i, g in df_dict.items():
#     print(i)
    if i == 'Open':
#         print('Yes')
#         print(max(g))
        highest_price = max(g.values())
        lowest_price = min(g.values())

print('The highest open price is ', highest_price)    
print('The lowest open price is ', lowest_price)  
```

```python
for k in df_dict['Open'].keys():
    if df_dict['Open'][k] == highest_price:
        highest_date = k
        
print('The highest open price is ', highest_price, ' on ', highest_date)  
```

```python
for k in df_dict['Open'].keys():
    if df_dict['Open'][k] == lowest_price:
        lowest_date = k
        
print('The lowest open price is ', lowest_price, ' on ', lowest_date) 
```
